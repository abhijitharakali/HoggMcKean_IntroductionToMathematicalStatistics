#### Exercise 5.1.5. 

Consider the R function *consistmean* which produces the plot shown in Figure $5.1.1$. Obtain a similar plot for the sample median when the distribution sampled is the $N(0,1)$ distribution. Compare the mean and median plots.


```python
def consistmean(ni,ne,dn):
    x =[]
    for n in range(ni,ne+1,dn):
        x.append(np.mean(np.random.normal(0,1,n)))
    return x
```

```python
ni = 10
ne = 2000
dn = 10

myrange = [n for n in range(ni,ne+1,dn)]
x = consistmean(ni,ne,dn)

plt.figure(figsize = (14,8))
plt.plot(myrange,x,linewidth=2,linestyle='', marker='o', color='r')
plt.grid()
plt.ylabel('Mean',fontsize=14,fontweight='bold')
plt.xlabel('Sample Size',fontsize=14,fontweight='bold')
plt.show()
```

```python
def consistmedian(ni,ne,dn):
    y =[]
    for n in range(ni,ne+1,dn):
        y.append(np.median(np.random.normal(0,1,n)))
    return y
```

```python
y = consistmedian(ni,ne,dn)

plt.figure(figsize = (14,8))
plt.plot(myrange,y,linewidth=2,linestyle='', marker='o', color='b')
plt.grid()
plt.ylabel('Median',fontsize=14,fontweight='bold')
plt.xlabel('Sample Size',fontsize=14,fontweight='bold')
plt.show()
```

#### Exercise 5.1.6. 

Write an R function that obtains a plot similar to Figure $5.1.1$ for the situation described in Example $5.1.2$. For the plot choose $\theta = 10$.

The "situation described in Example $5.1.2$" is not clear at all. Are we simulating the mean of the random variable $Y_n$ or are we simply simulating $Y_n$ itself and checking to see how it behaves as sample size is varied? I assumed the latter. Authors must have been clear about it.

```python
def consistmax_uniform(theta,ni,ne,dn):
    x =[]
    for n in range(ni,ne+1,dn):
        x.append(np.max(np.random.uniform(low=0.0, high=theta, size=n)))
    return x
```

```python
theta = 10
xu = consistmax_uniform(theta,ni,ne,dn)

plt.figure(figsize = (14,8))
plt.plot(myrange,xu,linewidth=2,linestyle='', marker='o', color='m')
plt.grid()
plt.ylabel('Max',fontsize=14,fontweight='bold')
plt.xlabel('Sample Size',fontsize=14,fontweight='bold')
plt.show()
```
